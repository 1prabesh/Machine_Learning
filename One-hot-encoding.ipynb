#### One hot encoding

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression as linreg
%matplotlib inline
```

```python
data=pd.read_excel('Documents\\device.xlsx')
data
```

```python
# gettig the dummys columns from pandas
dummy=pd.get_dummies(data.Device)
dummy
```

```python
# horizontal stacking of data and dummy
merg=pd.concat([data,dummy],axis='columns')
merg
```

```python
# dropping the dummy column
merged=merg.drop(['Device','Lenevo'],axis='columns')
merged
```

```python
#extracting feature X for training the model
X=merged.drop(['Price'],axis='columns')
X
```

```python
y=merged.Price
y
```

```python
reg=linreg()
reg.fit(X,y)
```

```python
reg.predict([[4233,1]])
```
